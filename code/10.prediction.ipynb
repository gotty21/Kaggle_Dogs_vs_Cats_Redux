```python
import numpy as np
import pandas as pd

import glob, cv2

from tensorflow import keras

from tqdm.notebook import tqdm
```

```python
model = keras.models.load_model("/content/drive/MyDrive/project/Kaggle_Dogs_vs_Cats_Redux/trained_model/test3_model.h5")
model.summary()
```

```python
paths = sorted(glob.glob("/content/datasets/test_resize_256/*"))

pred = []                  

for path in tqdm(paths):
    img = cv2.imread(path)
    img = np.array(img).astype('float32')/255

    img = img.reshape((1,) + img.shape)

    pred.append(model.predict(img)[0][0])

pred
```

```python
df = pd.DataFrame(paths, columns=['id'])
df['label'] = pred

id = []

for path in paths:
    id.append(int(path.split("/")[-1][:-4]))

df['id'] = id

df.sort_values('id').to_csv("/content/drive/MyDrive/project/Kaggle_Dogs_vs_Cats_Redux/submission.csv", index=False)
```
